```python
# importing the module
import pandas

# reading the files
f1 = pandas.read_excel("train_annotated.xlsx")
f2 = pandas.read_excel("Iclr 2020.xlsx")


```

```python
# merging the files
f3 = f1[["Paper_ID",
		"text"]].merge(f2[["Paper_ID",
										"R1_Rec", "R1_Length",
										"R1_Conf", "R2_Rec", "R2_Length",
										"R2_Conf", "R3_Rec", "R3_Length",
										"R3_Conf", "MR_Length", "Decision"]],
									on = "Paper_ID",
									how = "left")

# creating a new file
f3.to_excel("Iclr_2020.xlsx", index = False)
```
